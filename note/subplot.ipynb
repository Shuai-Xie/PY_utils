```python
# 为子图做 网格标签
x = range(5)
y = range(5)

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 10))

for row in axes:
    for col in row:
        col.plot(x, y)  # 作图

# 设置标签
# 取第 1 行设置 title
# axes[0, :]: 得到第1行的图，col 每一列起名字
for ax, col in zip(axes[0, :], ['1', '2', '3']):
    ax.set_title(col, size=15)

for ax, row in zip(axes[:, 0], ['A', 'B', 'C']):
    # rotation 旋转文字方向
    # labelpad 文字与轴的间距 pad 填充
    ax.set_ylabel(row, size=15, rotation=0, labelpad=15)

plt.show()
```

```python
# 坐标轴命名
x = range(5)
y = range(5)

# fig, ax = plt.subplots(nrows=2, ncols=2)

fig, ax = plt.subplots(nrows=2, ncols=2)

for row in ax:
    for col in row:
        col.plot(x, y)
        col.set_title('title')
        col.set_xlabel('x-axis')
        col.set_ylabel('x-axis')

# 自适应 子图间距
# Adjust subplot parameters to give specified padding.
fig.tight_layout()

plt.show()
```

```python
# 隐藏子图
x = range(10)
y = range(10)

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 10))

for cnt, ax in enumerate(axes.ravel()):  # 得到2维数字的每个元素
    if cnt < 7:
        ax.plot(x, y)
    else:
        ax.axis('off')  # hide subplot

plt.show()
```

```python
# 使用颜色
# Matplotlib 支持 RGB、'blue' 或 'b'、HTML十六进制

samples = range(1, 4)

for i, col in zip(samples, [(0.0, 0.0, 1.0), 'green', '#FF0000']):  # 颜色 list
    plt.plot([0, 10], [0, i], lw=3, color=col)

# 图例
plt.legend(['RGB values: (0.0, 0.0, 1.0)',
            "matplotlib names: 'green'",
            "HTML hex values: '#FF0000'"],
           loc='best')
plt.title('3 alternatives to define the color blue')

plt.show()
```

```python
# 颜色对应标签
# matlab color name
cols = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white']

samples = range(1, len(cols) + 1)

for i, col in zip(samples, cols):
    plt.plot([0, 10], [0, i], label=col, lw=3, color=col)

plt.legend(loc='best')
plt.title('matplotlib color names')

plt.show()
```

```python
# colormaps
from pylab import *
from numpy import *

a = outer(arange(0, 1, step=0.01), ones(10))  # 外积 (100, 10)

figure(figsize=(10, 5))
subplots_adjust(top=0.5, bottom=0.05, left=0.01, right=0.99)

maps = [m for m in cm.datad if not m.endswith('_r')]
maps.sort()
print maps

l = len(maps) + 1

for i, m in enumerate(maps):
    subplot(1, l, i + 1)  # 第 i+1 个 cmap 作图
    axis('off')
    imshow(a, aspect='auto', cmap=get_cmap(m), origin='lower')
    title(m, rotation=90, fontsize=10, pad=70)
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(14, 7))
samples = range(1, 16)  # [1,..,15]

# Default color cycle
for i in samples:
    ax0.plot([0, 10], [0, i], label=i, lw=3)

# Paired colormap
cmap = plt.get_cmap(name='Paired')
plt.gca().set_color_cycle([cmap(i) for i in np.linspace(0, 0.9, len(samples))])  # 在画 ax1 之前修改当前 axe

for i in samples:
    ax1.plot([0, 10], [0, i], label=i, lw=3)

# Annotation
ax0.set_title('Default color cycle')
ax1.set_title('Paired colormap')
ax0.legend(loc='upper left')
ax1.legend(loc='upper left')

plt.show()
```

```python
# 灰度图
plt.figure(figsize=(8, 6))

samples = np.arange(0, 1.1, 0.1)  # [0, 1.0]

for i in samples:
    plt.plot([0, 10], [0, i], label='gray-level %.1f' % i, lw=3,
             color=str(i))  # ! gray level has to be parsed as string

plt.legend(loc='upper left')
plt.title('gray-levels')

plt.show()
```

```python
# 散点图的边缘颜色 edgecolors
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

samples = np.random.randn(30, 2)

ax[0][0].scatter(samples[:, 0], samples[:, 1],
                 color='red',
                 label='color="red"')

ax[0][1].scatter(samples[:, 0], samples[:, 1],
                 edgecolor='white',  # default
                 c='red',
                 label='c="red", edgecolor="white"')

ax[1][0].scatter(samples[:, 0], samples[:, 1],
                 c='red',
                 label='c="red"')

ax[1][1].scatter(samples[:, 0], samples[:, 1],
                 edgecolor='0',
                 c='1',
                 label='color="1.0", edgecolor="0"')

for row in ax:
    for col in row:
        col.legend(loc='upper left')

plt.show()
```

```python
# 渐变样式
import matplotlib.pyplot as plt
import matplotlib.colors as col
import matplotlib.cm as cm
import numpy as np

# input data
mean_values = np.random.randint(1, 101, 100)
x_pos = range(len(mean_values))

fig = plt.figure(figsize=(20, 5))

# create colormap
cmap = cm.ScalarMappable(col.Normalize(min(mean_values),
                                       max(mean_values),
                                       cm.hot))

# plot bars
plt.subplot(131)  # 1行3个，第1个
plt.bar(x_pos, mean_values, align='center', alpha=0.5,
        color=cmap.to_rgba(mean_values))
plt.ylim(0, max(mean_values) * 1.1)

plt.subplot(132)
plt.bar(x_pos, np.sort(mean_values), align='center', alpha=0.5,  # np.sort 排序 使得 bar 有序增高
        color=cmap.to_rgba(mean_values))
plt.ylim(0, max(mean_values) * 1.1)

plt.subplot(133)
plt.bar(x_pos, np.sort(mean_values), align='center', alpha=0.5,
        color=cmap.to_rgba(np.sort(mean_values)))  # np.sort(mean_values) 使得 color 有序“增高”
plt.ylim(0, max(mean_values) * 1.1)

plt.show()
```

```python
# 标记 样式
import matplotlib.pyplot as plt

markers = [
    '.',  # point
    ',',  # pixel
    'o',  # circle
    'v',  # triangle down
    '^',  # triangle up
    '<',  # triangle_left
    '>',  # triangle_right
    '1',  # tri_down
    '2',  # tri_up
    '3',  # tri_left
    '4',  # tri_right
    '8',  # octagon
    's',  # square
    'p',  # pentagon
    '*',  # star
    'h',  # hexagon1
    'H',  # hexagon2
    '+',  # plus
    'x',  # x
    'D',  # diamond
    'd',  # thin_diamond
    '|',  # vline
]

plt.figure(figsize=(13, 10))
samples = range(len(markers))

for i in samples:
    plt.plot([i - 1, i, i + 1], [i, i, i], label=markers[i], marker=markers[i], markersize=10)

# Annotation
plt.title('Matplotlib Marker styles', fontsize=20)
plt.ylim([-1, len(markers) + 1])
plt.legend(loc='lower right')

plt.show()
```

```python
# 线 样式
import matplotlib.pyplot as plt

linestyles = ['-.', '--', 'None', '-', ':']

plt.figure(figsize=(8, 5))
samples = range(len(linestyles))

for i in samples:
    plt.plot([i - 1, i, i + 1], [i, i, i],
             label=linestyles[i],
             linestyle=linestyles[i],
             lw=4)

# Annotation
plt.title('Matplotlib line styles', fontsize=20)
plt.ylim([-1, len(linestyles) + 1])
plt.legend(loc='lower right')

plt.show()
```

```python
# 优化轴样式
import numpy as np
import math
import matplotlib.pyplot as plt

X = np.random.normal(loc=0.0, scale=1.0, size=300)
width = 0.5
bins = np.arange(math.floor(X.min()) - width,
                 math.ceil(X.max()) + width,
                 width)  # fixed bin size

ax = plt.subplot(111)

# remove axis at the top and to the right
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# hide axis ticks
plt.tick_params(axis="both", which="both", bottom="off", top="off",
                labelbottom="on", left="off", right="off", labelleft="on")

plt.hist(X, alpha=0.5, bins=bins)
plt.grid()

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')

plt.show()
```
